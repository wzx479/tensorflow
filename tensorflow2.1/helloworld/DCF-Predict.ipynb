```python
import time
import os
import tensorflow as tf
from tensorflow import keras
```

```python
os.environ["CUDA_VISIBLE_DEVICES"] = "0"  #GPU
#os.environ["CUDA_VISIBLE_DEVICES"] = "-1" #CPU

path =  "F:/PythonSources/SRC/data/dc_3000/val/1/"
img_path =  "./data/dc_3000/val/2/n01443537_18756.jpg"
save_model_dir = "./weight/dcf.h5"
```

```python
from models.vgg16 import VGG16
# model = VGG16(3)
# model.summary()
time_start=time.time()
model = keras.models.load_model(save_model_dir,compile=False)
time_end=time.time()
print('time cost',time_end-time_start,'s')
```

```python
time_start=time.time()
result = []
for i in range(500): 
    img_path = path+ 'dog.1'+ str(i).zfill(3) +'.jpg' 
#   print(img_path)
    image = tf.io.read_file(img_path)
    image = tf.image.decode_jpeg(image,channels=3)
    image = tf.image.resize(image,[224,224])
    image = tf.expand_dims(image,0)
    pred = model(image)
    result.append(tf.argmax(pred,axis=1))
time_end=time.time()
print('time cost',time_end-time_start,'s')

for i in result:
    print(i)
```
